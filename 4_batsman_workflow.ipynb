```python
from langgraph.graph import StateGraph, START, END
from typing import TypedDict
# from langgraph.graph.render import draw_mermaid_png
```

```python
class BatsmanState(TypedDict):
    
    runs = int
    balls = int
    fours = int
    sixes = int
    strike_rate = float
    bpb = float
    buoundery_percentage = float
    summary = str
```

```python
def calculate_sr(state: BatsmanState) -> BatsmanState:
    sr = (state['runs'] / state['balls']) * 100
    return {'sr': sr}
```

```python
def calculate_bpb(state: BatsmanState) -> BatsmanState:
    bpb = state['balls'] / (state['fours'] + state['sixes'])
    return {'bpb': bpb}
```

```python
def calculate_boundery_percentage(state: BatsmanState) -> BatsmanState:
    buoundery_percentage = (((state['fours'] * 4) + (state['sixes']* 6)) / state['balls']) * 100
    return {'buoundery_percentage': buoundery_percentage}
```

```python
def summery(state: BatsmanState) -> str:

    summery = f"""
    strike_rate: {state['sr']}\n
    bpb: {state['bpb']}
    buoundery_percentage: {state['buoundery_percentage']:.2f}%
    """

    return {'summary': summery}
```

```python
graph = StateGraph(BatsmanState)

graph.add_node('calculate_sr',calculate_sr)
graph.add_node('calculate_bpb',calculate_bpb)
graph.add_node('calculate_boundery_percentage',calculate_boundery_percentage)
graph.add_node('summery',summery)


#edges
graph.add_edge(START, 'calculate_sr')
graph.add_edge(START,'calculate_bpb')
graph.add_edge(START,'calculate_boundery_percentage')

graph.add_edge('calculate_sr','summery')
graph.add_edge('calculate_bpb','summery')
graph.add_edge('calculate_boundery_percentage','summery')


graph.add_edge('summery',END)


# graph_1 = graph.get_graph()

# graph.draw()

workflow = graph.compile()
```
